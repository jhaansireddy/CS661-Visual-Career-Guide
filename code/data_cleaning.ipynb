```python
mask=linkedin_job_postings.isnull()
print(linkedin_job_postings.isnull().sum())
```

```python
# Filling with dummy data since very few are missing
linkedin_job_postings["company"]=linkedin_job_postings["company"].fillna("Banco")
linkedin_job_postings["job_location"]=linkedin_job_postings["job_location"].fillna("Lake Park, IA")
```

```python
print(linkedin_job_postings.isnull().sum())
```

```python
linkedin_job_postings.drop_duplicates(subset=["company","job_link","last_processed_time","job_title","job_location","first_seen","search_city","search_country","search_position","job_level","job_type"],inplace=True)
```

```python
linkedin_job_postings["time"] = pd.to_datetime(linkedin_job_postings["last_processed_time"],dayfirst=True, format='mixed')
```

```python
linkedin_job_postings["dates"] = pd.to_datetime(linkedin_job_postings["last_processed_time"],dayfirst=True, format='mixed')
```

```python
linkedin_job_postings["current_time"]= pd.to_datetime(linkedin_job_postings["time"]).dt.time
```

```python
linkedin_job_postings["current_date"]= pd.to_datetime(linkedin_job_postings["dates"]).dt.date
```

```python
linkedin_job_postings.head()
```

```python
#deleting  unnesccesary column  for easy analysis
del linkedin_job_postings["got_summary"]
del linkedin_job_postings["got_ner"]
del linkedin_job_postings["is_being_worked"]
del linkedin_job_postings["last_processed_time"]
del linkedin_job_postings["dates"]
del linkedin_job_postings["time"]
```

```python
linkedin_job_postings["job_title"]=linkedin_job_postings["job_title"].str.lower().str.strip()
```

```python
linkedin_job_postings['job_location'] = linkedin_job_postings['job_location'].str.replace('  ',' ') # Replacing double spacing with one
linkedin_job_postings['job_location'] = linkedin_job_postings['job_location'].str.replace('   ',' ')
linkedin_job_postings['job_location'] = linkedin_job_postings['job_location'].str.replace(', ',',')
linkedin_job_postings['job_location'] = linkedin_job_postings['job_location'].str.replace(' ,',',')
linkedin_job_postings['job_location'] = linkedin_job_postings['job_location'].str.replace(' , ',',')
```

```python
# replacing all the values to get my data cleaned up a little bit
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('-',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('/',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('&',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('--',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('+',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('%',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('(',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace(')',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace(',',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('.',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('#',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('@',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace(';',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace(':',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('  ',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('   ',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('    ',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('     ',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('      ',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('  ',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('–',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('1',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('2',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('3',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('4',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('5',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('6',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('7',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('8',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('9',' ')
linkedin_job_postings['job_title'] = linkedin_job_postings['job_title'].str.replace('0',' ')
```
